```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# Setup Chrome WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

try:
    # Open the Trading Economics Stream page
    url = "https://tradingeconomics.com/stream"
    driver.get(url)

    # Wait until the news items load
    WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "list-group-item.te-stream-item"))
    )

    # Scroll to load more items
    for _ in range(5):
        driver.execute_script("window.scrollBy(0, 800);")
        time.sleep(1)

    # Locate all news items
    news_items = driver.find_elements(By.CLASS_NAME, "list-group-item.te-stream-item")

    # Extract data from each news item
    data = []
    for item in news_items:
        try:
            # Adjust selectors based on actual page structure
            time_stamp = item.find_element(By.CLASS_NAME, "te-time").text
            title = item.find_element(By.CLASS_NAME, "te-title").text
            details = item.find_element(By.CLASS_NAME, "te-description").text
            data.append([time_stamp, title, details])
        except Exception as e:
            print(f"Error extracting data for an item: {e}")
            continue

    # Create a DataFrame
    df = pd.DataFrame(data, columns=["Timestamp", "Title", "Details"])
    print(df)

    # Save to a CSV file
    df.to_csv("trading_economics_news.csv", index=False)
    print("Data saved to trading_economics_news.csv")

finally:
    # Close the driver
    driver.quit()
```
