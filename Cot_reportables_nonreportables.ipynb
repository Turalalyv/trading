```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pickle
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import re

# Setup Chrome WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open Facebook login page
driver.get("https://www.cftc.gov/dea/options/deacmelof.htm")
# input('enter')
pre_element = driver.find_element("tag name", "pre")
pre_text = pre_element.text

print(pre_text)
driver.quit()



```

```python
lines = pre_text.split("\n")

# Find all lines containing the phrase 'Traders in Financial Futures'
phrase = "Commitments of Traders with Delta-adjusted Options and Futures Combined"
matching_lines = []

tickers_line=[]
for idx, line in enumerate(lines):
    if phrase in line:
        matching_lines.append((idx, line))

# Output the results
if matching_lines:
    for line_idx, line in matching_lines:
        # print(f"Line {line_idx}: {line}")
        tickers_line.append(line_idx)
else:
    print(f"The phrase '{phrase}' was not found in the text.")
```

```python
html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <style>
        body { font-family: Arial, sans-serif; }
        .ticker-button { margin: 5px; padding: 10px; background-color: red; color: white; border: none; cursor: pointer; }
        .ticker-button:hover { background-color: darkred; }
        .ticker-data { display: none; margin-top: 20px; }
        table { width: 100%; border-collapse: collapse; }
        th, td { border: 1px solid #ddd; padding: 8px; text-align: center; }
        th { background-color: #f2f2f2; }
        .long-row { background-color: #d4edda; }
        .short-row { background-color: #f8d7da; }
        .change-positive { background-color: #d4edda; color: green; }
        .change-negative { background-color: #f8d7da; color: red; }
    </style>
    <script>
        function showData(tickerId) {
            document.querySelectorAll('.ticker-data').forEach(el => el.style.display = 'none');
            document.getElementById(tickerId).style.display = 'block';
        }
    </script>
</head>
<body>
    <div class="ticker-list">
"""

for i, ticker in enumerate(tickers):
    ticker_name = list(ticker.keys())[0]
    html_content += f'<button class="ticker-button" onclick="showData(\'ticker-{i}\')">{ticker_name}</button>'

for i, ticker in enumerate(tickers):
    ticker_name = list(ticker.keys())[0]
    data = ticker[ticker_name]

    html_content += f'<div id="ticker-{i}" class="ticker-data">'

    open_interest = data['Open Interest']
    html_content += f"""
    <h3>Open Interest</h3>
    <table>
        <tr><th>Value</th><th>Change</th><th>Percent</th></tr>
        <tr><td>{open_interest['Value']}</td><td>{open_interest['Change']}</td><td>{open_interest['Percent']}%</td></tr>
    </table>
    """

    html_content += f"""
    <h3>Non-Commercial vs Commercial</h3>
    <table>
        <tr><th>Type</th><th>Non-Commercial</th><th>Non-Comm Change</th><th>Commercial</th><th>Comm Change</th></tr>
    """
    for key in ['Long', 'Short', 'Spreading']:
        # Check if 'Spreading' exists in both Non-Commercial and Commercial
        non_comm = data['Non-Commercial'].get(key, None)
        comm = data['Commercial'].get(key, None)

        if non_comm is not None and comm is not None:
            row_class = "long-row" if key == "Long" else "short-row"
            
            non_change_class = "change-positive" if non_comm['Change'] > 0 else "change-negative"
            comm_change_class = "change-positive" if comm['Change'] > 0 else "change-negative"

            html_content += f"""
            <tr class="{row_class}">
                <td>{key}</td>
                <td>{non_comm['Value']} ({non_comm['Percent']}%)</td>
                <td class="{non_change_class}">{non_comm['Change']}</td>
                <td>{comm['Value']} ({comm['Percent']}%)</td>
                <td class="{comm_change_class}">{comm['Change']}</td>
            </tr>
            """
    
    # Add a separate row for Net Interest after Long and Short rows
    long_value = data['Non-Commercial'].get('Long', {'Value': 0})['Value']
    short_value = data['Non-Commercial'].get('Short', {'Value': 0})['Value']
    net_interest = long_value - short_value

    html_content += f"""
    <tr>
        <td><strong>Net Interest</strong></td>
        <td colspan="4"><strong>{net_interest}</strong></td>
    </tr>
    """

    html_content += "</table></div>"

html_content += """
    </div>
</body>
</html>
"""

with open("tickers_display.html", "w") as file:
    file.write(html_content)

print("HTML file created: tickers_display.html")
```
