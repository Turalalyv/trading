```python
import yfinance as yf
import numpy as np
import pandas as pd
import pandas_ta as ta 

# Function to check the trend status based on EMA
def get_status(row):
    if (row['ema50'] < row['close'] and
        row['ema200'] < row['close'] and
        row['ema50'] > row['ema200']):
        return 'bullish'
    elif (row['ema50'] > row['close'] and
        row['ema200'] > row['close'] and
        row['ema50'] < row['ema200']):
        return 'bearish'
    else:
        return 'hold'

# Function to calculate EMA and trend status for a single ticker
def get_uptrends(ticker):
    df = yf.download(ticker, start='2000-01-01', end='2024-12-31', interval='1d')
    df = df.tail(500)  # Take the last 500 rows for analysis
    
    # Clean the DataFrame
    df.columns = df.columns.str.lower()
    columns_to_drop = ['stock splits', 'dividends']
    columns_to_drop = [col for col in columns_to_drop if col in df.columns]
    if columns_to_drop:
        df.drop(columns_to_drop, axis=1, inplace=True)
    
    df.reset_index(inplace=True)  # Reset index to get 'date' as a column
    df.columns = df.columns.str.lower()
    
    df['date'] = df['date'].dt.tz_localize('UTC')  
    df['date'] = df['date'].dt.tz_convert(None)
    
    # Calculate EMA indicators
    df['ema50'] = ta.ema(df['close'], length=50)
    df['ema200'] = ta.ema(df['close'], length=200)
    
    # Ensure ema200 is float64
    df['ema200'] = df['ema200'].astype(np.float64)
    
    # Calculate ADX and other indicators
    df.ta.adx(length=14, append=True)
    
    # Drop rows with NaN values
    df.dropna(inplace=True)
    
    # Get the second last row to evaluate the most recent date
    last_row = df.iloc[-2]
    status = get_status(last_row)
    
    # Return the trend status and relevant data
    return {
        'ticker': ticker,
        'date': last_row['date'],
        'close': last_row['close'],
        'ema50': last_row['ema50'],
        'ema200': last_row['ema200'],
        'adx': last_row['ADX_14'],
        'status': status
    }

def get_all_ticker_status(ticker_list):
    # Create an empty DataFrame to store the results
    results_df = pd.DataFrame(columns=['ticker', 'date', 'close', 'ema50', 'ema200', 'adx', 'status'])
    

    for ticker in ticker_list:
        result = get_uptrends(ticker)
        results_df = pd.concat([results_df, pd.DataFrame([result])], ignore_index=True)
    
    return results_df

# Define the list of tickers
ticker_list = [
    'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'BRK-B', 'META', 'TSLA', 
    'UNH', 'XOM', 'JNJ', 'V', 'PG', 'JPM', 'MA', 'LLY', 'CVX', 'HD', 
    'PFE', 'ABBV', 'MRK', 'PEP', 'KO', 'AVGO', 'COST', 'DIS', 'CMCSA', 
    'CSCO', 'INTC', 'CRM', 'TXN', 'TMO', 'BMY', 'VZ', 'NKE', 'ORCL', 
    'NEE', 'MCD', 'ADBE', 'PM', 'T', 'ABT', 'AMD', 'UNP', 'MDT', 'CVS', 
    'QCOM', 'HON', 'GS'
]

final_results = get_all_ticker_status(ticker_list)


```

```python
import yfinance as yf
import pandas as pd

# Function to get volume data and calculate the 14-day mean
def get_volume_data(ticker):
    # Download stock data for the last 500 days
    df = yf.download(ticker, start='2000-01-01', end='2024-12-31', interval='1d')
    df = df.tail(500)  # Analyze the last 500 days
    
    # Clean the DataFrame
    df.columns = df.columns.str.lower()
    columns_to_drop = ['stock splits', 'dividends']
    columns_to_drop = [col for col in columns_to_drop if col in df.columns]
    if columns_to_drop:
        df.drop(columns_to_drop, axis=1, inplace=True)
    
    df.reset_index(inplace=True)  # Reset index to get 'date' as a column
    df.columns = df.columns.str.lower()

    df['date'] = df['date'].dt.tz_localize('UTC')  
    df['date'] = df['date'].dt.tz_convert(None)
    
    # Ensure that we have enough data to calculate the 14-day mean
    if len(df) < 14:
        return None

    # Calculate the 14-day mean volume
    df['14_day_mean_volume'] = df['volume'].rolling(window=14).mean()

    # Get the last two days of data along with the mean
    df = df[['date', 'volume', '14_day_mean_volume']].tail(3)  # Keep the last 3 rows for comparison

    # Add ticker information to the DataFrame
    df['ticker'] = ticker
    return df

# Function to get volume data for a list of tickers
def get_all_volume_data(ticker_list):
    all_volume_data = []
    
    # Loop over each ticker and collect volume data
    for ticker in ticker_list:
        volume_data = get_volume_data(ticker)
        if volume_data is not None:
            all_volume_data.append(volume_data)
    
    return pd.concat(all_volume_data, ignore_index=True)

# Define the list of tickers
ticker_list = [
    'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'BRK-B', 'META', 'TSLA', 
    'UNH', 'XOM', 'JNJ', 'V', 'PG', 'JPM', 'MA', 'LLY', 'CVX', 'HD', 
    'PFE', 'ABBV', 'MRK', 'PEP', 'KO', 'AVGO', 'COST', 'DIS', 'CMCSA', 
    'CSCO', 'INTC', 'CRM', 'TXN', 'TMO', 'BMY', 'VZ', 'NKE', 'ORCL', 
    'NEE', 'MCD', 'ADBE', 'PM', 'T', 'ABT', 'AMD', 'UNP', 'MDT', 'CVS', 
    'QCOM', 'HON', 'GS'
]

# Get the results for volume data
all_volume_data_df = get_all_volume_data(ticker_list)

# Display the DataFrame
print(all_volume_data_df)
```
