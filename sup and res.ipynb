```python
import pandas as pd 
import numpy as np 
import datetime as dt 
import yfinance as yf
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
df=yf.Ticker("NQ=F")
df=df.history(period='max')
df.reset_index(inplace=True)
df.drop(['Dividends','Stock Splits'],axis=1,inplace=True)
df=df[df['Volume']!=0]
df.columns=df.columns.str.lower()
df=df[df['date']>'2023-01-01']
df
```

```python
def support(df1, l, n1, n2):
    for i in range(l - n1 + 1, l + 1):
        if df1.low.iloc[i] > df1.low.iloc[i - 1]:
            return 0
    for i in range(l + 1, l + n2 + 1):
        if df1.low.iloc[i] < df1.low.iloc[i - 1]:
            return 0
    return 1

def resistance(df1, l, n1, n2):
    for i in range(l - n1 + 1, l + 1):
        if df1.high.iloc[i] < df1.high.iloc[i - 1]:
            return 0
    for i in range(l + 1, l + n2 + 1):
        if df1.high.iloc[i] > df1.high.iloc[i - 1]:
            return 0
    return 1

sr = []
n1 = 3
n2 = 2

for row in range(n1, 205):
    if support(df, row, n1, n2):
        sr.append((row, df.low.iloc[row], 1))
    if resistance(df, row, n1, n2):
        sr.append((row, df.high.iloc[row], 2))

print(sr)
```

```python
s=0
e=len(df)
dfpl=df[s:e]
fig=go.Figure(data=[go.Candlestick(x=dfpl.index,
                                  open=dfpl['open'],
                                  close=dfpl['close'],
                                  low=dfpl['low'],
                                  high=dfpl['high'])])
c=0
for line in sr:
    fig.add_shape(type='line', x0=dfpl.index[s], y0=line[1],
                  x1=dfpl.index[e - 1], y1=line[1])
fig.update_layout(title='Us100')
fig.show()
```

```python
plotlist1=[x[1] for x in sr if x[2]==1]
plotlist2=[x[1] for x in sr if x[2]==2]
plotlist1.sort()
plotlist2.sort()
for i in range(1,len(plotlist1)):
    if(i>=len(plotlist1)):
        break
    if abs(plotlist1[i]-plotlist1[i-1]<=0.005):
        plotlist1.pop(i)
for i in range(1,len(plotlist2)):
    if(i>=len(plotlist2)):
        break
    if abs(plotlist2[i]-plotlist2[i-1]<=0.005):
        plotlist2.pop(i)
s=0
e=len(df)
dfpl=df[s:e]
fig=go.Figure(data=[go.Candlestick(x=dfpl.index,
                                  open=dfpl['open'],
                                  close=dfpl['close'],
                                  low=dfpl['low'],
                                  high=dfpl['high'])])
c=0
while (1):
    if (c>len(plotlist1)-1):
        break
    fig.add_shape(type='line',x0=s,y0=plotlist1[c],
                 x1=e,
                 y1=plotlist1[c],
                 line=dict(color='MediumPurple',width=1))
    c+=1
while (1):
    if (c>len(plotlist2)-1):
        break
    fig.add_shape(type='line',x0=s,y0=plotlist2[c],
                 x1=e,
                 y1=plotlist2[c],
                 line=dict(color='Blue',width=1))
    c+=1    
fig.show()
```
