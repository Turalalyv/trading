```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pickle
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
import re

# Setup Chrome WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open Facebook login page
driver.get("https://www.cftc.gov/dea/futures/financial_lf.htm")
# input('enter')
pre_element = driver.find_element("tag name", "pre")
pre_text = pre_element.text

print(pre_text)
driver.quit()


```

```python
import pandas as pd



# Splitting the data into lines
lines = pre_text.split("\n")

# Debugging step to inspect the structure
print("Lines in pre_text:")
for idx, line in enumerate(lines):
    print(f"Line {idx}: {line}")

# Extracting metadata
contract_line = lines[3]
contract_name = contract_line.split("  (")[0].strip()
cftc_code = lines[4].split("#")[-1].split()[0]
open_interest = int(lines[4].split("is")[-1].strip().replace(",", ""))

# Extracting positions
positions_line = lines[6].strip()
positions = list(map(int, positions_line.split()))

# Extracting changes
changes_line = lines[8].strip()
changes = list(map(int, changes_line.split()))

# Extracting percentages
percentages_line = lines[10].strip()
percentages = list(map(float, percentages_line.split()))

# Extracting number of traders
traders_line = lines[12].strip()
num_traders = [
    int(x) for x in traders_line.split() if x.isdigit()
]  # Extract numbers, skipping dots

# Creating DataFrame
columns = [
    "Dealer_Intermediary_Long",
    "Dealer_Intermediary_Short",
    "Dealer_Intermediary_Spreading",
    "Asset_Manager_Long",
    "Asset_Manager_Short",
    "Asset_Manager_Spreading",
    "Leveraged_Funds_Long",
    "Leveraged_Funds_Short",
    "Leveraged_Funds_Spreading",
    "Other_Reportables_Long",
    "Other_Reportables_Short",
    "Other_Reportables_Spreading",
    "Nonreportable_Positions_Long",
    "Nonreportable_Positions_Short",
]
df = pd.DataFrame(
    [positions, changes, percentages],
    columns=columns,
    index=["Positions", "Changes", "Percentages"],
).T

# Adding metadata
metadata = {
    "Contract Name": contract_name,
    "CFTC Code": cftc_code,
    "Open Interest": open_interest,
    "Total Traders": num_traders[0],
}
metadata_df = pd.DataFrame([metadata])

# Display results
print("\nMetadata:")
print(metadata_df)
print("\nParsed DataFrame:")
print(df)
```
